## Why Data Visualization?

Instead of reading large tables or numbers, visualizations make it easier to:

- **Spot patterns and trends**
- **Compare values quickly**
- **Explain results clearly**

---

## Common Types of Visualizations

- **Bar Charts** – Compare categories  
- **Line Plots** – Show changes over time  
- **Scatter Plots** – Show relationships between two variables  
- **Histograms** – Show data distribution  
- **Heatmaps** – Show intensity or correlation

<h2 style="color:#FFB84C; font-family:'Georgia', serif;">What Are Subplots?</h2>
<p style="color:#D3D3D3; font-family:Arial, sans-serif;">
Subplots allow you to <b>display multiple plots</b> in a single figure window.  
You can arrange them in a grid (rows × columns) using <code>plt.subplot()</code> or <code>plt.subplots()</code>.
</p>



```python
plt.subplot(2,2,1)
plt.plot(y,x, color ='skyblue',marker='^', linestyle='--')

plt.subplot(2,2,4)
plt.plot(x,y, color ='red',marker='o', linestyle='--')
```

<h2 style="color:#FFB84C; font-family:'Georgia', serif;">Overview</h2>
<p style="color:#D3D3D3; font-family:Arial, sans-serif;">
In the <b>Object-Oriented (OO) approach</b>, you create a <b>Figure</b> and one or more <b>Axes</b> objects,  
and then call methods on those objects to customize and draw your plots.<br><br>
This approach offers <b>greater control and flexibility</b> than the state-based <code>plt</code> interface.
</p>

---

<div style="height:2px; background-color:#6CA6C1; margin:20px 0;"></div>

<h2 style="color:#FFB84C; font-family:'Georgia', serif;">Basic Example</h2>


```python
# Create one Figure and one Axes
fig, ax = plt.subplots()

# Create a twin y-axis sharing the same x-axis
ax2 = ax.twinx()

# Plot data on each Axes
ax.plot([1, 2, 3, 4], [2, 4, 6, 8], color='skyblue', marker='o')
ax2.plot([1, 2, 3, 4], [9, 5, 6, 8], color='orange', marker='x')

# Set labels and title
ax.set_title("Simple OO Plot", color='white')
ax.set_xlabel("X-Axis")
ax.set_ylabel("Left Y-Axis")
ax2.set_ylabel("Right Y-Axis")

plt.show()
```

```python
fig = plt.figure(figsize=(8,5),dpi=100 )
# figsize = (width, height) in inches
# dpi = dots per inch (image resolution)
ax = fig.add_axes([0.1,0.1 ,0.8,0.8])
ax.plot(x, [i**2 for i in x], color='skyblue', marker='o')

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('x vs x²')


ax2 = fig.add_axes([0.6,0.2 ,0.5,0.5])
ax2.plot(x, x, color='red', marker='o')

```

<h2 style="color:#FFB84C; font-family: 'Georgia', serif;">1️⃣ Line Plot</h2>
<p style="color:#D3D3D3; font-family: Arial, sans-serif;">
Used to visualize trends over time or continuous data.<br>
Example: <i>Stock prices, temperature changes, accuracy vs. epochs.</i>
</p>

---

<div style="height: 2px; background-color:#6CA6C1; margin: 20px 0;"></div>

<h2 style="color:#FFB84C; font-family: 'Georgia', serif;">2️⃣ Bar Plot</h2>
<p style="color:#D3D3D3; font-family: Arial, sans-serif;">
Displays categorical data with rectangular bars.<br>
Example: <i>Sales by region, counts of labels, frequency of words.</i>
</p>

---

<div style="height: 2px; background-color:#6CA6C1; margin: 20px 0;"></div>

<h2 style="color:#FFB84C; font-family: 'Georgia', serif;">3️⃣ Histogram</h2>
<p style="color:#D3D3D3; font-family: Arial, sans-serif;">
Shows distribution of a numerical variable by grouping into bins.<br>
Example: <i>Age distribution, pixel intensity, word length.</i>
</p>

---

<div style="height: 2px; background-color:#6CA6C1; margin: 20px 0;"></div>

<h2 style="color:#FFB84C; font-family: 'Georgia', serif;">4️⃣ Scatter Plot</h2>
<p style="color:#D3D3D3; font-family: Arial, sans-serif;">
Displays relationship between two continuous variables.<br>
Example: <i>Height vs. Weight, Income vs. Spending.</i>
</p>

---

<div style="height: 2px; background-color:#6CA6C1; margin: 20px 0;"></div>

<h2 style="color:#FFB84C; font-family: 'Georgia', serif;">5️⃣ Box Plot</h2>
<p style="color:#D3D3D3; font-family: Arial, sans-serif;">
Summarizes distribution using quartiles, median, and outliers.<br>
Example: <i>Compare salary ranges or model errors.</i>
</p>

---

<div style="height: 2px; background-color:#6CA6C1; margin: 20px 0;"></div>

<h2 style="color:#FFB84C; font-family: 'Georgia', serif;">6️⃣ Pie Chart</h2>
<p style="color:#D3D3D3; font-family: Arial, sans-serif;">
Represents parts of a whole as circular slices.<br>
Example: <i>Market share, topic distribution, sentiment proportions.</i>
</p>

---

<div style="height: 2px; background-color:#6CA6C1; margin: 20px 0;"></div>

<h2 style="color:#FFB84C; font-family: 'Georgia', serif;">7️⃣ Heatmap</h2>
<p style="color:#D3D3D3; font-family: Arial, sans-serif;">
Shows matrix-style data with color intensity.<br>
Example: <i>Correlation matrices, confusion matrices.</i>
