<h2 style="color:#FFB84C; font-family: 'Georgia', serif;">1️⃣ distplot / histplot</h2>
<p style="color:#D3D3D3; font-family: Arial, sans-serif;">
Used to visualize the distribution of a dataset using a histogram and an optional KDE (Kernel Density Estimate).<br>
<b>Note:</b> <i><code>distplot</code></i> is deprecated; use <i><code>histplot</code></i> instead.<br>
Example use:<br>
<i><code>sns.histplot(data, kde=True)</code></i><br>
Shows both the frequency of data and its estimated probability density.
</p>

---

<div style="height: 2px; background-color:#6CA6C1; margin: 20px 0;"></div>

<h2 style="color:#FFB84C; font-family: 'Georgia', serif;">2️⃣ jointplot</h2>
<p style="color:#D3D3D3; font-family: Arial, sans-serif;">
Displays a bivariate relationship between two variables along with their individual distributions.<br>
Example use:<br>
<i><code>sns.jointplot(x="height", y="weight", data=df, kind="scatter")</code></i><br>
Includes both a scatter plot and histograms/KDE plots on the axes.
</p>

---

<div style="height: 2px; background-color:#6CA6C1; margin: 20px 0;"></div>

<h2 style="color:#FFB84C; font-family: 'Georgia', serif;">3️⃣ pairplot</h2>
<p style="color:#D3D3D3; font-family: Arial, sans-serif;">
Plots pairwise relationships between multiple variables in a dataset.<br>
Each variable is paired with every other variable to show correlations.<br>
Example use:<br>
<i><code>sns.pairplot(df)</code></i><br>
Diagonal subplots show distributions, while off-diagonals show scatter relationships.
</p>

---

<div style="height: 2px; background-color:#6CA6C1; margin: 20px 0;"></div>

<h2 style="color:#FFB84C; font-family: 'Georgia', serif;">4️⃣ rugplot</h2>
<p style="color:#D3D3D3; font-family: Arial, sans-serif;">
Displays individual data points along an axis — like small vertical ticks (“rugs”) that indicate the data distribution.<br>
Example use:<br>
<i><code>sns.rugplot(data=df["age"])</code></i><br>
Useful for visualizing raw data density on top of other plots (e.g., histograms or KDEs).
</p>

```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

We use distribution plots to visualize the distribution of quantitative data.

Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

- distplot/histplot
- jointplot
- pairplot
- rugplot

```python
df =sns.load_dataset('tips',cache=True, data_home=None)
```

```python
df.head()
```

```python
df['size'].unique()  # HENCE THIS COL IS ALSO QUALITATIVE 
```

<div style="text-align: center;">
<h3 style="color:#B0C4DE; font-family: 'Trebuchet MS', sans-serif; font-size: 32px;">
distplot / histplot
</h3>
</div>
<div style="height: 2px; background-color:#6CA6C1; margin: 20px 0;"></div>

```python
#sns.displot(df['total_bill'] , bins=20 , kde = True)
# or
#sns.histplot(df['total_bill'] , bins=20)


plt.subplot(1,2 ,1)
sns.histplot(df['total_bill'] , bins=20 , kde = True)

plt.subplot(1,2 ,2)
sns.histplot(df['tip'] , bins=20 , kde = True)
```

<div style="text-align: center;">
<h3 style="color:#B0C4DE; font-family: 'Trebuchet MS', sans-serif; font-size: 32px;">
 jointplot
</h3>
</div>
<div style="height: 2px; background-color:#6CA6C1; margin: 20px 0;"></div>

```python
sns.jointplot(x='total_bill', y='tip', data=df , kind = 'scatter')
#kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }
```

```python
sns.jointplot(x='total_bill', y='tip', data=df , kind = 'reg' )
```
