<h2 style="color:#FFC55C; font-family:'Georgia',serif;">1️⃣ boxplot</h2>
<p style="color:#A4C8E1; font-family:Arial,sans-serif;">
Displays the distribution of data based on five summary statistics: minimum, first quartile (Q1), median, third quartile (Q3), and maximum.<br>
Great for comparing distributions across categories.<br>
Example:<br>
<i><code>sns.boxplot(x="category", y="value", data=df)</code></i><br>
Highlights outliers and the spread of data for each category.
</p>

---

<div style="height:2px; background-color:#6FA8DC; margin:20px 0;"></div>

<h2 style="color:#FFC55C; font-family:'Georgia',serif;">2️⃣ violinplot</h2>
<p style="color:#A4C8E1; font-family:Arial,sans-serif;">
Combines the boxplot with a kernel density estimate (KDE) to show the full data distribution shape.<br>
Useful for visualizing multimodal distributions.<br>
Example:<br>
<i><code>sns.violinplot(x="category", y="value", data=df, inner="quartile")</code></i>
</p>

---

<div style="height:2px; background-color:#6FA8DC; margin:20px 0;"></div>

<h2 style="color:#FFC55C; font-family:'Georgia',serif;">3️⃣ stripplot</h2>
<p style="color:#A4C8E1; font-family:Arial,sans-serif;">
Plots individual data points for each category with slight jitter for better visibility.<br>
Best for small datasets to show all observations.<br>
Example:<br>
<i><code>sns.stripplot(x="category", y="value", data=df, jitter=True)</code></i>
</p>

---

<div style="height:2px; background-color:#6FA8DC; margin:20px 0;"></div>

<h2 style="color:#FFC55C; font-family:'Georgia',serif;">4️⃣ swarmplot</h2>
<p style="color:#A4C8E1; font-family:Arial,sans-serif;">
Similar to <code>stripplot</code> but automatically arranges points to avoid overlap.<br>
It provides a better visual representation of the distribution of data points.<br>
Example:<br>
<i><code>sns.swarmplot(x="category", y="value", data=df)</code></i>
</p>

---

<div style="height:2px; background-color:#6FA8DC; margin:20px 0;"></div>

<h2 style="color:#FFC55C; font-family:'Georgia',serif;">5️⃣ barplot</h2>
<p style="color:#A4C8E1; font-family:Arial,sans-serif;">
Shows the mean (or another estimator like median) of a numerical variable for each category, with confidence intervals.<br>
Ideal for summarizing central tendency.<br>
Example:<br>
<i><code>sns.barplot(x="category", y="value", data=df, ci=95)</code></i>
</p>

---

<div style="height:2px; background-color:#6FA8DC; margin:20px 0;"></div>

<h2 style="color:#FFC55C; font-family:'Georgia',serif;">6️⃣ countplot</h2>
<p style="color:#A4C8E1; font-family:Arial,sans-serif;">
Displays the count (frequency) of each category.<br>
Best for categorical data without a numeric variable.<br>
Example:<br>
<i><code>sns.countplot(x="category", data=df)</code></i><br>
Simple and effective for showing class balance or frequency distribution.
</p>

## Categorical Data Plots
Now let's discuss using seaborn to plot categorical data! :

- boxplot
- violinplot
- stripplot
- swarmplot
- barplot
- countplot

Let's go through examples of each!

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
df.head()
```

<div style="text-align: center;">
<h3 style="color:#B0C4DE; font-family: 'Trebuchet MS', sans-serif; font-size: 32px;">
barplot
</h3>
</div>
<div style="height: 2px; background-color:#6CA6C1; margin: 20px 0;"></div>

```python
sns.barplot(x = df['sex'],y = df['tip'])
```

You can change the estimator object to your own function, that converts a vector to a scalar 
- Mean (default): np.mean 
- Median: np.median 
- Sum: np.sum 
- Count: len (to count the number of observations) 
- Standard Deviation: np.std 
- Minimum: np.min 
- Maximum: np.max

```python
import numpy as np 
sns.barplot(x = df['sex'],y = df['tip'] ,estimator= np.sum)
```

## boxplot and violinplot
boxplots and violinplots are used to shown the distribution of categorical data. A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range.

<div style="text-align: center;">
<h3 style="color:#B0C4DE; font-family: 'Trebuchet MS', sans-serif; font-size: 32px;">
violinplot
</h3>
</div>
<div style="height: 2px; background-color:#6CA6C1; margin: 20px 0;"></div>

```python
sns.violinplot(x='day' , y='total_bill' , data=df)
```

## stripplot and swarmplot
The stripplot will draw a scatterplot where one variable is categorical. A strip plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.

The swarmplot is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they don’t overlap. This gives a better representation of the distribution of values, although it does not scale as well to large numbers of observations (both in terms of the ability to show all the points and in terms of the computation needed to arrange them).

<div style="text-align: center;">
<h3 style="color:#B0C4DE; font-family: 'Trebuchet MS', sans-serif; font-size: 32px;">
swarmplot
</h3>
</div>
<div style="height: 2px; background-color:#6CA6C1; margin: 20px 0;"></div>

```python
sns.swarmplot(x='day' , y= 'total_bill' , data= df)
```

```python
sns.violinplot(x= 'tip',y = 'day',data = df)
sns.swarmplot(x= 'tip',y = 'day',data = df)
```
